```python
import json                                                                                                                                                                                                                                 
meta = json.load(open('/home/yifeng/aic_xvla_data/cableholder-phase-combined-xvla/aic_meta.json'))                                                                                                                                          
instructions = {}                                                                                                                                                                                                                           
for item in meta['datalist']:                                                                                                                                                                                                               
    ins = item.get('instruction', '(missing)')                                                                                                                                                                                              
    instructions[ins] = instructions.get(ins, 0) + 1                                                                                                                                                                                        
for ins, c in sorted(instructions.items(), key=lambda x: -x[1]):                                                                                                                                                                            
    print(f'  [{c:4d}] {ins}')                                                                                                                                                                                                              
```
